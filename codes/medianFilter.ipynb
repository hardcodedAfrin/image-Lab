```python
import pandas as pd
import numpy as np
from PIL import Image
```

```python
img = Image.open('images/panda.jpeg').convert('L').resize((640 // 2,426 // 2))
img
```

```python
imgMat = np.array(img)
imgMat
```

```python
size = 5
np.linspace(- (size // 2), size // 2, size).astype(int)
```

```python
size = 5
def getMedian(i,j,mat):
    p = np.linspace(- (size // 2), size // 2, size).astype(int)
    arr = []
    for dx in p:
        if i + dx >= 0 and i + dx < len(mat):
            for dy in p:
                if j + dy >= 0 and j + dy < len(mat[0]):
                    arr.append(mat[i+dx][j+dy]) #append the value of the pixel at (i+dx, j+dy) to the array
    return np.median(arr) #return the median value of the array
```

```python
medMat = imgMat.copy()
for i in range(len(imgMat)):
    for j in range(len(imgMat[0])):
        medMat[i][j] = getMedian(i,j,imgMat)
```

```python
final = Image.fromarray(medMat)
```
